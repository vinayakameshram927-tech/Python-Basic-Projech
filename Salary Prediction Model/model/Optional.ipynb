```python
x = df.drop('Salary',axis = 1)
y = df['Salary']
```

```python
x.shape
```

```python
y.shape
```

```python
cols = df.select_dtypes(include = 'object').columns
```

```python
encoded_x = pd.get_dummies(x, columns = cols, drop_first = True)
```

```python
encoded_x.head()
```

```python
encoded_x = encoded_x.astype(int)
```

```python
encoded_x.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x = pd.DataFrame(scaler.fit_transform(encoded_x), columns = encoded_x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled_x,y,test_size=0.2,random_state=101)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr_model = lr.fit(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

```python
print(y_pred[0:10:1])
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
