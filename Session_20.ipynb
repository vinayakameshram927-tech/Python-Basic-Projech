```python
# Independent Feature:
x = df.drop('price', axis = 1)
# dependent Feature :
y = df['price']
```

```python
x.shape
```

```python
y.shape
```

```python
print("Data set Before One hot Encoding :")
df.head(5)
```

```python
encoded_x = pd.get_dummies(x, drop_first=True)
encoded_x = encoded_x.astype(int)
```

```python
print("Data set After One hot Encoding :")
encoded_x.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x = pd.DataFrame(scaler.fit_transform(encoded_x), columns=encoded_x.columns)
X = scaled_x
```

```python
print("Scaled Dataset: ")
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
print("Size After Training: \n")
print("X_train: ",X_train.shape)
print("X_test : ",x_test.shape)
print("Y_train: ",y_train.shape)
print("X_test : ",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr_model = lr.fit(X_train,y_train)
print("Coefficient: \n",lr_model.coef_)
```

```python
print("Intercept: \n",lr_model.intercept_)
```

```python
y_pred = lr_model.predict(x_test)
print("Predicted Values: ")
y_pred[0:11:1]
```

```python
print("Actual Values: ")
y_test[0:11:1]
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

the model explains 83.10% of the variation, indicating good performance.

```python
import joblib
joblib.dump(lr_model,'LR_ford_car.pkl')
joblib.dump(scaler,'scaler.pkl')
joblib.dump(encoded_x.columns,'columns.pkl')
```
